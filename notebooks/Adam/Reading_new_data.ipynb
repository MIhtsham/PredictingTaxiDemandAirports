```python
df_list = []
for i in date_list:
    data_temp = pd.read_excel('../../raw_data/new_data/lot_info.xlsx', 
    #                          skiprows=11, 
    #                          skip_footer=2, 
                              sheet_name = i,
                              header=None, 
                              names=['Hour', 
                                     'P7_num', 
                                     'P7_percent', 
                                     'DH_num', 
                                     'DH_percent', 
                                     'Term_A_num', 
                                     'Term_A_percent', 
                                     'Total_percent', 
                                     'MAT', 
                                     'Term_B_Pax_waittime', 
                                     'Term_C_Pax_waittime', 
                                     'None_1', 
                                     'Term_D_Pax_waittime', 
                                     'None_2'])
    data_temp = data_temp.loc[data_temp.Hour.isin(hours), good_cols]
    data_temp.insert(loc = 0, column='Date', value = i)
    df_list.append(data_temp)
```

```python
final_data = pd.concat(df_list)
```

```python
final_data
```

```python
final_data.Hour = final_data.Hour.apply(lambda row: row.split('-')[0])
```

```python
pd.to_datetime('11.19.17 0600', format = '%m.%d.%y %H%M')
```

```python
final_data['dt'] = final_data["Date"].map(str) + ' ' + final_data["Hour"]
```

```python
final_data = final_data.set_index('dt')
```

```python
final_data.index = pd.to_datetime(final_data.index, format = '%m.%d.%y %H%M')
```

```python
final_data = final_data.drop(['Date', 'Hour'], axis=1)
```

```python
final_data.to_csv('../../clean_data/hourly_lot_status.csv')
```
